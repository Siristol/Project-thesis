```python
# CIFAR-10
from keras.datasets import cifar10
import matplotlib.pyplot as plt
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer','dog', 'frog', 'horse', 'ship', 'truck']
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

plt.figure(figsize=(12, 12))
for i in range(25): # Display 25 samples
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i])
    # Make labels more visible with title instead of xlabel
    plt.title(f'{class_names[y_train[i][0]]}', fontsize=12, fontweight='bold', color='black')
    
plt.tight_layout()
plt.show()
```

```python
# MNIST
from keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
(x_train, y_train),(x_test, y_test) = mnist.load_data()
indexes = np.random.randint(0, x_train.shape[0], size=25)
images = x_train[indexes]
labels = y_train[indexes]

# Alternative: Print labels above images
plt.figure(figsize=(8,8))
for i in range(len(indexes)):
    plt.subplot(5, 5, i + 1)
    image = images[i]
    plt.imshow(image, cmap='gray')
    # Put label at the top of the image
    plt.text(0.5, 1.05, f'{labels[i]}', transform=plt.gca().transAxes, 
             ha='center', va='bottom', fontsize=14, fontweight='bold')
    plt.axis('off')
    
plt.tight_layout()
plt.show()
```
